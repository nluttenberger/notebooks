## Building the adjacency matrix

```python
import io
import os
import re
from xml.dom import minidom
import json
import codecs
import networkx as nx
from networkx.algorithms import bipartite
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
import locale
locale.setlocale(locale.LC_ALL, 'de-DE.utf-8')
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

#### Read json file

```python
os.chdir('C:/Users/nlutt/Documents/Websites/graphLab/sampleSpaces/compareHD-YO/graphs')
cwd = os.getcwd()
#with open('combinedGraph.json', 'r', encoding='utf-8') as f:
with open('simpleGraph.json', 'r', encoding='utf-8') as f:
  combinedGraph = json.load(f)
recipes = combinedGraph.get('recipes')
print (len(recipes))
#for recipe in recipes:
    #print (recipe)
```

#### Building the pandas dataframe
Rows stand for ingredients, columns stand for recipes

```python
left = pd.DataFrame(index = [])
for recipe in recipes:
    right = pd.DataFrame(index = recipe.get('ingredients'), data = {recipe.get('recipeName'):1})
    left = pd.concat([left, right], axis=1).fillna(0).astype(int).sort_index(axis=0)   
print (left)
print (left.T)
```

#### Build bipartite graph from pandas dataframe and project
top holds recipe names  
bottom holds ingredient names

```python
B = nx.Graph()
top = []
bottom = []
e_list = []
for recipe in recipes:
    top.append(recipe.get('recipeName'))
for recipe in recipes:
    for ingredient in recipe.get('ingredients'):
        if not(ingredient in bottom):
            bottom.append(ingredient)
B.add_nodes_from(top, bipartite=0)
B.add_nodes_from(bottom, bipartite=1)
for recipe in recipes:
    for ingredient in recipe.get('ingredients'):
        nm = recipe.get('recipeName')
        e_list.append((nm,ingredient))
B.add_edges_from(e_list)
G = bipartite.weighted_projected_graph(B, bottom)
df = pd.DataFrame (nx.to_numpy_array(G, nodelist=bottom),columns=bottom,index=bottom,dtype=np.int8)
print(df.to_string(col_space=13))
```

#### Set node attributes: i-name, i-class, occ(urrence)

```python
os.chdir('C:/Users/nlutt/myPyPro/second')
cwd = os.getcwd()
with open('igt_cat.json', encoding='utf-8') as file:
    cat_data = json.load(file)
attr_dict = {igt: cat_data[igt] for igt in bottom}
#print (attr_dict)
occ_list = [igt for rcp in recipes for igt in rcp.get('ingredients')]
occ_dict = Counter(occ_list)
occ_attr = {k:{'occ':occ_dict.get(k)} for k in occ_dict.keys()}
#print (occ_attr)
nx.set_node_attributes(G, attr_dict)
nx.set_node_attributes(G, occ_attr)
#print (list(G.nodes(data=True)))
```

#### Set node attribute sub(graph)

```python
Arecipes = [rcp.get('recipeName') for rcp in recipes if rcp.get('recipeName') in combinedGraph.get('collections')[0].get('A').get('recipes')]
Brecipes = [rcp.get('recipeName') for rcp in recipes if rcp.get('recipeName') in combinedGraph.get('collections')[1].get('B').get('recipes')]
print ('#A recipes: ', str(len(Arecipes)).rjust(5))
print ('#B recipes: ', str(len(Brecipes)).rjust(5))
Aingredients = set([igt for rcp in recipes for igt in rcp.get('ingredients') if rcp.get('recipeName') in Arecipes])
Bingredients = set([igt for rcp in recipes for igt in rcp.get('ingredients') if rcp.get('recipeName') in Brecipes])
ABingredients = Aingredients.intersection(Bingredients)
Aingredients_pure = Aingredients.difference(ABingredients)
Bingredients_pure = Bingredients.difference(ABingredients)
Asub_dict = {igt: {'sub':'A'} for igt in Aingredients_pure}
Bsub_dict = {igt: {'sub':'B'} for igt in Bingredients_pure}
ABsub_dict = {igt: {'sub':'AB'} for igt in ABingredients}
print ('#A nodes: ', len(Aingredients_pure))
#print (Aingredients)
print ('#B nodes: ', len(Bingredients_pure))
#print (Bingredients)
print ('#AB nodes: ', len(ABingredients))
#print (ABingredients)
sub_dict = {**Asub_dict, **Bsub_dict, **ABsub_dict}
nx.set_node_attributes(G, sub_dict)
#for n in list(G.nodes(data=True)):
    #print (n)
```

#### Set edge attribute: id

```python
e_attr = {}
for e in list(G.edges(data=True)):
    x = [e[0],e[1]]
    x.sort(key=locale.strxfrm)
    id = str(x[0]) + '--' + str(x[1])
    xx = (e[0],e[1])
    e_attr[xx] = {'id':id}
nx.set_edge_attributes(G, e_attr)
#for e in list(G.edges(data=True)):
    #print (e)
```

#### Set edge attribute: sub(graph)

```python
A_attr = {(e[0],e[1]):{'sub': 'A'} for e in list(G.edges(data=True)) if (e[0] in Aingredients_pure and e[1] in Aingredients_pure) or (e[0] in Aingredients_pure and e[1] in ABingredients) or (e[0] in ABingredients and e[1] in Aingredients_pure)}
B_attr = {(e[0],e[1]):{'sub': 'B'} for e in list(G.edges(data=True)) if (e[0] in Bingredients_pure and e[1] in Bingredients_pure) or (e[0] in Bingredients_pure and e[1] in ABingredients) or (e[0] in ABingredients and e[1] in Bingredients_pure)}
AB_attr = {(e[0],e[1]):{'sub': 'AB'} for e in list(G.edges(data=True)) if e[0] in ABingredients and e[1] in ABingredients}
e_attr = {**A_attr,**B_attr,**AB_attr}
print ('#A edges: ',len(A_attr),'\n#B edges: ',len(B_attr),'\n#AB edges: ',len(AB_attr))
nx.set_edge_attributes(G, e_attr)
#for e in list(G.edges(data=True)):
    #print (e)
```

#### Generate edge sets

```python
A_pure  = [(e[0],e[1]) for e in list(G.edges(data=True)) if (e[0] in Aingredients_pure  and e[1] in Aingredients_pure)]
A_mixed = [(e[0],e[1]) for e in list(G.edges(data=True)) if (e[0] in Aingredients_pure  and e[1] in ABingredients) or (e[0] in ABingredients and e[1] in Aingredients_pure)]
AB_set  = [(e[0],e[1]) for e in list(G.edges(data=True)) if  e[0] in ABingredients and e[1] in ABingredients]
B_pure  = [(e[0],e[1]) for e in list(G.edges(data=True)) if (e[0] in Bingredients_pure  and e[1] in Bingredients_pure)]
B_mixed = [(e[0],e[1]) for e in list(G.edges(data=True)) if (e[0] in Bingredients_pure  and e[1] in ABingredients) or (e[0] in ABingredients and e[1] in Bingredients_pure)]
#print (len(A_pure),  A_pure)
#print (len(A_mixed), A_mixed)
#print (len(AB_set),  AB_set)
#print (len(B_pure),  B_pure)
#print (len(B_mixed), B_mixed)
print (len(A_pure))
print (len(A_mixed))
print (len(AB_set))
print (len(B_pure))
print (len(B_mixed))
```

#### Compute Krackhardt's E-I-index

```python
def Krack (EL,IL):
    return (EL-IL)/(EL+IL)
EL_A = len(A_mixed) + len(AB_set)
IL_A = len(A_pure)
print ('A edges: ', Krack(EL_A,IL_A))
EL_B = len(B_mixed) + len(AB_set)
IL_B = len(B_pure)
print ('B edges: ', Krack(EL_B,IL_B))
xx = [n for n in Aingredients_pure for neigh in nx.all_neighbors(G,n) if neigh in Aingredients_pure]
IL_nodes_A = Counter(xx)
xx = [n for n in Aingredients_pure for neigh in nx.all_neighbors(G,n) if neigh in ABingredients]
EL_nodes_A = Counter(xx)

xx = [n for n in Bingredients_pure for neigh in nx.all_neighbors(G,n) if neigh in Bingredients_pure]
IL_nodes_B = Counter(xx)
xx = [n for n in Bingredients_pure for neigh in nx.all_neighbors(G,n) if neigh in ABingredients]
EL_nodes_B = Counter(xx)
#print (IL_nodes_A)
#print (EL_nodes_A)
```

#### Compute Krackhardts's E-I index for nodes and export data to excel

```python
i = [n for n in Aingredients_pure]
k = []
for n in Aingredients_pure:
    x = (EL_nodes_A.get(n) if EL_nodes_A.get(n) != None else 0)
    y = (IL_nodes_A.get(n) if IL_nodes_A.get(n) != None else 0)
    z = round(Krack(x,y),2)
    k.append(z)
A_idx = pd.Series (k, name='E-I index for A', index=i)

i = [n for n in Bingredients_pure]
k = []
for n in Bingredients_pure:
    x = (EL_nodes_B.get(n) if EL_nodes_B.get(n) != None else 0)
    y = (IL_nodes_B.get(n) if IL_nodes_B.get(n) != None else 0)
    z = round(Krack(x,y),2)
    k.append(z)
B_idx = pd.Series (k, name='E-I index for B', index=i)

os.chdir('C:/Users/nlutt/Documents/Websites/graphLab/sampleSpaces/compareHD-YO/graphs')
with pd.ExcelWriter('out.xlsx') as writer: 
    A_idx.to_excel(writer, sheet_name='E-I index Davidis')
    B_idx.to_excel(writer, sheet_name='E-I index Ottolenghi')
```

#### Compute relations from AB vegetable nodes to A or B spice/herb nodes

#### Convert G to graphviz dot language notation

```python
dot = 'graph {\ngraph[rankdir="LR", outputorder="edgesfirst"]\nnode[fontname="Arial", fontsize=120, shape=circle, style=filled, fixedsize=shape];\n'
for u,v,att in G.edges(data=True):
    dot += u+' -- '+v+' [penwidth='+str(att.get('weight'))
    if att.get('weight') > 1:
        dot += ', color=Red]\n'
    else:
        dot += ']\n'
for u,att in G.nodes(data=True):
    dot += u+' [width=' + str(8+6*np.log(att.get('occ'))) + ', label=' + str(att.get('i-name')) + ', class=' + str(att.get('i-class')) + ']\n'
dot += '}'
fn = 'xx.dot'
os.chdir('C:/Users/nlutt/Documents/Websites/graphLab/sampleSpaces/compareHD-YO/graphs')
with codecs.open(fn, 'w', encoding = 'utf8') as file:
    file.write(dot)
#print (dot)
```
